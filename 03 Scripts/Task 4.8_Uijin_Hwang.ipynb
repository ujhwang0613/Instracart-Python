#### Q6. The team now wants to target different types of spenders in their marketing campaigns. This can be achieved by looking at the prices of the items people are buying. Create a spending flag for each user based on the average price across all their orders using the following criteria (a) If the mean of the prices of products purchased by a user is lower than 10, then flag them as a Low spender.(b) If the mean of the prices of products purchased by a user is higher than or equal to 10, then flag them as a High spender.

```python
# Create new column 'avg_price'

ords_prods_merge['avg_price'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
ords_prods_merge.head(15)
```

```python
# Create a flag that assigns a types of spenders label to a user ID based on average prices.

ords_prods_merge.loc[ords_prods_merge['avg_price'] < 10, 'spender_flag'] = 'Low spender'
```

```python
ords_prods_merge.loc[ords_prods_merge['avg_price'] >= 10, 'spender_flag'] = 'High spender'
```

```python
ords_prods_merge['spender_flag'].value_counts(dropna = False)
```

#### Q7. In order to send relevant notifications to users within the app (for instance, asking users if they want to buy the same item again), the Instacart team wants you to determine frequent versus non-frequent customers. Create an order frequency flag that marks the regularity of a users ordering behavior according to the median in the days_since_prior_order column.

```python
ords_prods_merge['median_days_since_prior_order'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.mean)
```

```python
ords_prods_merge.loc[ords_prods_merge['median_days_since_prior_order'] >20, 'frequency_flag'] = 'Non-frequent customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['median_days_since_prior_order'] >10) & (ords_prods_merge['median_days_since_prior_order'] <= 20), 'frequency_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['median_days_since_prior_order'] <=10, 'frequency_flag'] = 'Frequent customer'
```

```python
ords_prods_merge['frequency_flag'].value_counts(dropna = False)
```
